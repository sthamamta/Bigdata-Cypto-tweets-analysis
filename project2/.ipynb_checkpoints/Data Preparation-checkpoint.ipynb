## Impute missing value with mode

```python
# since all the missing values columns are categorical, therefore, we impute them with mode

data_train = data_train.apply(lambda x:x.fillna(x.value_counts().index[0]))

data_test = data_test.apply(lambda x:x.fillna(x.value_counts().index[0]))
```

```python
numeric_lst = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numerical_cols = list(data_train.select_dtypes(include=numeric_lst).columns)
categorical_cols = list(set(data_train.columns)- set(numerical_cols))
```

country feature has a lot of unique values, therefore, one hot encoding of country feature will cause increase in dimension of dataset

```python
data_train['country'].value_counts().plot(kind='bar')
plt.show()
```

most of country feature has value United-states so we map the country feature value United-States to 1 and all other values to 0

```python
data_train['country'] = [1 if i == 'United-States'  else 0 for i in data_train['country']]
data_test['country'] = [1 if i == 'United-States'  else 0 for i in data_test['country']]
```

## Dropping redundant column

```python
# dropping redundant column education
data_train = data_train.drop(['education'], axis=1)
data_test = data_test.drop(['education'], axis=1)
```

## Encoding categorical columns

```python
# handling categorical columns
# 1. Country (binary)
# 2. sex (binary)
# 3. target(binary)
# 4. education(ordinal)
# 5. workclass (one-hot)
# 6. relationship(one-hot)
# 7. race(label)
# 8. marital-status (onehot)
# 9. occupation (label)

one_hot_cols = ['workclass','relationship','marital_status']
label_encode_cols =['occupation','sex','race']

data_train['target'] = data_train.target.map({'<=50K': 0, '>50K': 1})
data_test['target'] = [0 if i == '<=50K'  else 1 for i in data_test['target']]
```

```python
# label encoding
le = LabelEncoder()

for col in label_encode_cols:
    le.fit(data_train[col])
    # Transform both training and testing data
    data_train[col] = le.transform(data_train[col])
    data_test[col] = le.transform(data_test[col])
```

```python
#one hot encoding

def one_hot_encode(df, cols):
    dummies = pd.get_dummies(df[cols])
    enc = pd.concat([df, dummies], axis=1)
    enc = enc.drop(cols, axis=1)
    return(enc)

data_train = one_hot_encode(data_train,one_hot_cols)
data_test  = one_hot_encode(data_test, one_hot_cols)

# add missing attributes
for attributes in data_train.keys():
    if attributes not in data_test.keys():
        print("Adding missing feature {}".format(attributes))
        data_test[attributes] = 0
```
